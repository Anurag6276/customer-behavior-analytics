```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('customer_shopping_behavior.csv')
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df = df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
df.columns
```

```python
sns.countplot(x="category", data=df)
plt.xticks(rotation=45)
plt.title("Number of Purchases by Category")
plt.show()
```

<span style="color:blue">From this countplot we can say most of the purchases are of Clothing category followed by Accessories"</span>

```python
plt.hist(df["purchase_amount"])
plt.title("Distribution of Purchase Amount")
plt.xlabel("Purchase Amount (USD)")
plt.ylabel("Number of Customers")
plt.show()
```

<span style="color:blue">From this histogram we can say purchase amount are mostly uniformly distributed in different ranges,indicating that customers spend fairly evenly without a dominant spending range.</span>

```python
#Create a column age_group
labels = ["Young Adult", "Adult", "Middle-aged", "Senior"]
df["age_group"]=pd.qcut(df["age"], q=4, labels = labels)
```

```python
df[["age_group","age"]].head(15)
```

```python
#create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[["purchase_frequency_days","frequency_of_purchases"]].head(15)
```

```python
df[["discount_applied","promo_code_used"]].head(10)
```

<span style="color:blue">Now lets find if both columns are useful or we can remove any redunadant columns because both give same information</span>

```python
(df["discount_applied"]==df["promo_code_used"]).all()
```

<span style="color:blue">One column is redundant because presence of both do not add any new information so lets remove one</span>

```python
df = df.drop("promo_code_used", axis=1)
```

```python
df.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x="category", y="purchase_amount", data=df)
plt.xticks(rotation=45)
plt.title("Purchase Amount by Category")
plt.xlabel("Category")
plt.ylabel("Purchase Amount (USD)")
plt.show()
```

<span style="color:blue">Final insights from the cleaned data using boxplot:<br>
Purchase behavior is fairly consistent across all categories, with similar median spending.<br>Outerwear shows slightly lower spending, while all categories exhibit a wide range of purchase values.<br>
This indicates a balanced business with diverse customer spending patterns and no extreme outliers.</span>
